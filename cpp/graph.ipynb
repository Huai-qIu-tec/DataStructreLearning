```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(1, 10, 10)
y1 = x * 2
y2 = x ** 2
plt.plot(x, y1, y2)
```

```python
plt.plot(x, y1, color='r', marker='o', linestyle='--', linewidth=1, alpha=0.65)
plt.plot(x, y2, color='b', marker='s', linestyle='-', linewidth=1, alpha=0.65)
plt.legend(['$2x$', '$x^2$'])
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$x^2$')
```

```python
relation_x = np.random.random((5, 5))
relation_x
```

```python
relation_x = pd.DataFrame(data=relation_x, columns=['x1', 'x2', 'x3', 'x4', 'x5'], index=['x1', 'x2', 'x3', 'x4', 'x5'])
corr = np.corrcoef(relation_x)
corr
```

```python
sns.heatmap(corr, square=True)
```

```python
sns.set_theme(style="whitegrid",font='TFlux Regular',font_scale=1)
sns.heatmap(data=corr, vmin=-1, vmax=1, cmap="YlGn", annot=True, linewidths=3)
```
